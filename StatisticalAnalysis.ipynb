```python
# Correct p-vals

def p_adjust_bh(p):
    """Benjamini-Hochberg p-value correction for multiple hypothesis testing.
    Got this from: https://stackoverflow.com/questions/7450957/how-to-implement-rs-p-adjust-in-python """
    p = np.asfarray(p)
    by_descend = p.argsort()[::-1]
    by_orig = by_descend.argsort()
    steps = float(len(p)) / np.arange(len(p), 0, -1)
    q = np.minimum(1, np.minimum.accumulate(steps * p[by_descend]))
    h = q<0.05
    return h,q[by_orig]

p_eff_Da= np.load('correlations/Person_eff_Da.npy')
p_SW_Da= np.load('correlations/Person_SW_Da.npy')

file_pi2 = open('correlations/betweenness_D.pickle',"rb")
betcent_Da = pickle.load(file_pi2)
file_pi2.close()

h1_betcent_Da, p_betcent_Da= p_adjust_bh(np.load('correlations/P_value_betw_Da.npy'))
r_betcent_Da= np.load('correlations/Person_betcent_Da.npy')

file_pi2 = open('correlations/betweenness_D_chronic.pickle',"rb")
betcent_Dc = pickle.load(file_pi2)
file_pi2.close()

h1_betcent_Dc, p_betcent_Dc= p_adjust_bh(np.load('correlations/P_value_betw_Dc.npy'))
r_betcent_Dc= np.load('correlations/Person_betcent_Dc.npy')



file_pi2 = open('correlations/closeness_D.pickle',"rb")
closcent_Da = pickle.load(file_pi2)
file_pi2.close()

h1_closcent_Da, p_closcent_Da= p_adjust_bh(np.load('correlations/P_value_clos_Da.npy'))
r_closcent_Da= np.load('correlations/Person_closcent_Da.npy')

file_pi2 = open('correlations/closeness_D_chronic.pickle',"rb")
closecent_Dc = pickle.load(file_pi2)
file_pi2.close()

h1_closcent_Dc, p_closcent_Dc= p_adjust_bh(np.load('correlations/P_value_clos_Dc.npy'))
r_closcent_Dc= np.load('correlations/Person_closcent_Dc.npy')

```

```python
#  Correlation Betweeen Centrality measures and Clinical Scores 

meanBet = np.zeros(63)
medianBet = np.zeros(63)
meanClose = np.zeros(63)
medianClose = np.zeros(63)

for i,k in enumerate(betcent_Da.keys()):
    betcent = np.array(list(betcent_Da[k].values()))
    meanBet[i] = np.mean(betcent[np.where(betcent!=0)[0]])
    
    closecent = np.array(list(closcent_Da[k].values()))
    meanClose[i] = np.mean(closecent[np.where(closecent!=0)[0]])

r = np.zeros((2,len(scores.columns)))
p = np.zeros((2,len(scores.columns)))
p_corrected = np.zeros((2,len(scores.columns)))
for i,score in enumerate(scores.columns):
    r[0,i], p[0,i] = scipy.stats.pearsonr(x=meanBet,y=scores[score])
    r[1,i], p[1,i] = scipy.stats.pearsonr(x=meanClose,y=scores[score])


for i in range(2):
    p_corrected[i,:] = p_adjust_bh(p[i,:])[1]  
h_corrected = p_corrected<0.05

fig,axis = plt.subplots(2,6,figsize=(30,10))
count = 0
for i,score in enumerate(scores.columns):
    if i not in [2,5,6]:
        if h_corrected[0,i]:
            axis[0,count].scatter(scores[score],meanBet,c='g')        
            
        else:
            axis[0,count].scatter(scores[score],meanBet,c='r')

        axis[0,count].set_title('r= %0.2g pval =%0.4g'%(r[0,i],p_corrected[0,i]),fontsize=20)
        
        if h_corrected[1,i]:
            axis[1,count].scatter(scores[score],meanClose,c='g')
        else:
            axis[1,count].scatter(scores[score],meanClose,c='r')
        
        axis[1,count].set_title('r= %0.2g pval =%0.4g'%(r[1,i],p_corrected[1,i]),fontsize=20)
        axis[1,count].set_xlabel(score,fontsize=20)
        count = count+1
        axis[0,0].set_ylabel('Avg. Betweenness\nCentrality',fontsize=20)
        axis[1,0].set_ylabel('Avg. Closeness\nCentrality',fontsize=20)
```
